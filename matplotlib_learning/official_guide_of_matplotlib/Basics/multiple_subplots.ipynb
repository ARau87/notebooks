```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Creating test data to plot
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
# Plot the data into two separate plots

# Params: n_rows n_cols index_of_current_plot
plt.subplot(2,1,1)
plt.plot(x1, y1, 'o-')
plt.title('Plot 1')
plt.ylabel('Damped oscillation')

plt.subplot(2,1,2)
plt.plot(x2, y2, '.-')
plt.title('Plot 2')
plt.xlabel('time (s)')
plt.ylabel('Undamped oscillation')

plt.show()
```
